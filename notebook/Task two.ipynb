```python
import talib as ta
import pandas as pd
import matplotlib.pyplot as plt
```

# SMI RSI and MACD analysis of each stock

## APPLE STOCK

```python
stock_aapl = pd.read_csv(r"C:\Users\Ramlla\Desktop\Financial\data\AAPL_historical_data.csv")
stock_aapl.head()
```

```python
stock_aapl = stock_aapl[['Date', 'Open', 'High', 'Low', 'Adj Close', 'Volume']]
stock_aapl['Date'] = pd.to_datetime(stock_aapl['Date'])  # Ensure Date column is in datetime format
stock_aapl.set_index('Date', inplace=True)  # Set Date as index
```

Calculating simple moving average, relative strength index and Moving average convergence divergence

```python
#SMA  is calculated across 50 
#RSI is calculated over 14 day period 
# sets the short-term EMA (Exponential Moving Average) to 12 days
# sets the long-term EMA to 26 days
# sets the Signal line, which is the EMA of the MACD line, to 9 days

stock_aapl['SMA_50'] = ta.SMA(stock_aapl['Adj Close'], timeperiod=50)
stock_aapl['RSI'] = ta.RSI(stock_aapl['Adj Close'], timeperiod=14)
stock_aapl['MACD'], stock_aapl['MACD_Signal'], stock_aapl['MACD_Hist'] = ta.MACD(stock_aapl['Adj Close'],  fastperiod=12, slowperiod=26, signalperiod=9)


plt.figure(figsize=(12, 6))
plt.plot(stock_aapl.index, stock_aapl['RSI'], label='RSI', color='purple')
plt.axhline(70, linestyle='--', color='red', label='Overbought Threshold (70)')
plt.axhline(30, linestyle='--', color='green', label='Oversold Threshold (30)')
plt.xlabel('Date')
plt.ylabel('RSI')
plt.title('Relative Strength Index (RSI)')
plt.legend()
plt.savefig('RSI_plot_appl.png') 
plt.show()


plt.figure(figsize=(12, 6))
plt.plot(stock_aapl.index, stock_aapl['MACD'], label='MACD Line', color='blue')
plt.plot(stock_aapl.index, stock_aapl['MACD_Signal'], label='MACD Signal Line', color='red')
plt.bar(stock_aapl.index, stock_aapl['MACD_Hist'], label='MACD Histogram', color='gray', alpha=0.3)
plt.xlabel('Date')
plt.ylabel('MACD')
plt.title('MACD and Signal Line')
plt.legend()
plt.savefig('MACD and Signal Line of appl.png') 
plt.show()

```

The stock spent less time in oversold conditions (RSI < 30). This suggests that there were fewer periods of significant price drops relative to the stock's recent price history. As for the MACD Pre-2005 stability indicates balanced trends and low momentum.Post-2020 volatility reflects a more dynamic market environment with significant momentum changes.High positive values Up to 8 indicate strong buying interest while low values show down to -6 shows strong selling pressure.

```python
print(stock_aapl.head())
```

```python
# Plot Closing Price and SMA
plt.figure(figsize=(12,6))
plt.plot(stock_aapl.index, stock_aapl['Adj Close'], label=' Adjusted Close Price')
plt.plot(stock_aapl.index, stock_aapl['SMA_50'], label='50-Day SMA')
plt.legend()
plt.savefig('closing and sma of appl.png')
plt.show()
```

Pre 2005 the stock price was generally in line with its 50-day average, suggesting a balanced market condition during that time. After 2013 it suggests that the stock started to experience more significant fluctuations and trends. After year 2020 it shows the stock has been experiencing substantial growth.In year 2024 the stock price stays above the 50-Day SMA, showing ongoing positive momentum and strong investor confidence.
